```python
#import packages
import pandas as pd
import pandas_profiling as pp
```

```python
#loading dataset of ipl matches
matches=pd.read_excel("C:\\Users\\Admin\\IPL Data Analysis\\Initial Datasets\\IPL Matches(2008-2021).xlsx")
```

```python
#Pandas Profiling before Data Preprocessing
#pre profiling of matches dataset
pre_profile =pp.ProfileReport(matches) 
pre_profile.to_file(output_file="matches_profiling_before_preprocessing.html")
```

```python
#Filling in the missing values of city column
#find the venues corresponding to which the values of city are empty
matches[matches.city.isnull()][['city','venue']].venue.unique()
```

```python
#Sharjah Cricket Stadium in Abu Dhabi
#Dubai International Cricket Stadium in Dubai
for i in matches.venue:
    if i=="Sharjah Cricket Stadium":
        matches['city']=matches['city'].fillna('Abu Dhabi')
    elif i=="Dubai International Cricket Stadium":
        matches['city']=matches['city'].fillna('Dubai')
```

```python
#find total null values
matches.isnull().sum()
```

```python
#dismissal_kind and player_dismissed can have null values since each delivery is considered
#Player of match has 3 missing values. 
#Let's see this based on result column to find why it is empty

matches[matches.player_of_match.isnull()][['player_of_match','result']]
#The results shows when there was no result(may be match was abandoned) then there was no player_of_match which makes sense
```

```python
matches['team1'].unique()
```

```python
#For simplification we replace these team names into recent one---11 teams from 2008-2021
# Replace the ipl team 1 name since different names exist for one team
matches['team1']=matches['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
matches['team1']=matches['team1'].str.replace('Punjab Kings','Kings XI Punjab')
matches['team1']=matches['team1'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
matches['team1']=matches['team1'].str.replace('Pune Warriors','Rising Pune Supergiant')
matches['team1']=matches['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
# Replace the ipl team 2 name since different names exist for one team
matches['team2']=matches['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
matches['team2']=matches['team2'].str.replace('Punjab Kings','Kings XI Punjab')
matches['team2']=matches['team2'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
matches['team2']=matches['team2'].str.replace('Pune Warriors','Rising Pune Supergiant')
matches['team2']=matches['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#Replace the ipl team names since different names exist for one team
matches['winner']=matches['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
matches['winner']=matches['winner'].str.replace('Punjab Kings','Kings XI Punjab')
matches['winner']=matches['winner'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
matches['winner']=matches['winner'].str.replace('Pune Warriors','Rising Pune Supergiant')
matches['winner']=matches['winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
#Replace the ipl team names since different names exist for one team
matches['toss_winner']=matches['toss_winner'].str.replace('Delhi Daredevils','Delhi Capitals')
matches['toss_winner']=matches['toss_winner'].str.replace('Punjab Kings','Kings XI Punjab')
matches['toss_winner']=matches['toss_winner'].str.replace('Rising Pune Supergiants','Rising Pune Supergiant')
matches['toss_winner']=matches['toss_winner'].str.replace('Pune Warriors','Rising Pune Supergiant')
matches['toss_winner']=matches['toss_winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
matches.isnull().sum()
```

```python
#to fetch set of stadiums
matches.venue.unique()
```

```python
#replace Wankhede Stadium
matches['venue']=matches['venue'].str.replace('Wankhede Stadium, Mumbai, Mumbai','Wankhede Stadium, Mumbai')
```

```python
#to fetch year from date
matches['date'] = pd.to_datetime(matches['date'])
matches['year'] = matches['date'].dt.year
matches
```

```python
#let us make over & ball into single column over
b1["ball"] = b1["ball"]-1
b1["over"] = b1["over"].astype(str) +"."+ b1["ball"].astype(str)
b1.head()
```

```python
#delete uncommon columns from b1 dataset
b1=b1.drop(columns=["ball","non_boundary","is_wicket","extras_type"])
```

```python
#delete uncommon columns from b2 dataset
b2=b2.drop(columns=["Unnamed: 14","Unnamed: 15","Unnamed: 16","Unnamed: 17","Unnamed: 18","Unnamed: 19","Unnamed: 20"])
```

```python
#convert over datatype from object to float64
b1['over']=b1['over'].astype('float64')
```

```python
#inning is considered as object type in b2 so convert to int
b2['inning']=b2['inning'].str.replace('2nd innings','2')
b2['inning']=b2['inning'].str.replace('1st innings','1')
b2['inning']=b2['inning'].astype('int64')
```
